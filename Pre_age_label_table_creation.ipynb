```python
#NORMALIZE_NAME: the function takes the subject path basename and splits it according to the underscore
#in order to take the first string and use it as index in dataframe.
#If a string is repeated, the function adds the first letter of the second string after the underscore

def normalize_name(subject_path, add_name):
    #Components= []
    subject = os.path.basename(subject_path)
    subject = re.sub(r'^(DE|D|DI|LO|DEL)_', '', subject)
    components =  subject.split("_")
    if add_name[0]==components[0] or add_name[1]==components[0] or add_name[2]==components[0] :
            return components[0].title() + components[1][0].capitalize()
    else:
            return components[0].title()
        
    
#GET_SUBJECT_METADATA:the function recives, as input, the subject path and the list of name repeated already known
#the output are the path base name (SURNAME_NAME) and the normalized name
#add_name needs to contain the list of name repeated


def get_subject_metadata(subject_path, subjects_with_name=[]):
    dirname = os.path.basename(subject_path)
    return (dirname, normalize_name(subject_path, add_name=[name for name in subjects_with_name]))
```

```python
def age_path_label_table(par_labels,age,label):
    df_par = pd.read_csv(par_labels, sep=";", header=None,skipinitialspace=True)
    df_par[0]= df_par[0].str.replace("Carlantonio", "Carlatonio", case = False) 
    df_par[0] = df_par[0].str.strip()
    df_age = pd.read_csv(age, sep=";", header=None,skipinitialspace=True)
    df_age[0] = df_age[0].str.strip()
    next_df_par=df_par[0]
    df_diff = pd.concat([df_par[0],df_age[0]]).drop_duplicates(keep=False)
    for i in range(len(df_diff.index)):
        df_age=df_age.drop([df_diff.index[i]])
    df_par.columns = ["Subject", f"{label}"]
    df_par = df_par.set_index('Subject')
    df_age.columns = ["Subject", "Age"]
    df_age = df_age.set_index('Subject')
    result = df_age.join(df_par, on='Subject')
    
    Df_Subjects_dirs = []
    for subject_path in glob.glob(f"{datasetDir}/*"):
        if subject_path.endswith(".csv"):
            continue
        if subject_path.endswith(".xlsx"):
            continue
        if subject_path.endswith("CSV_DATA"):
            continue
        subjects_path = [get_subject_metadata(subject_path , subjects_with_name=[SUBJECT_NAME for SUBJECT_NAME in SUBJECT_NAME_REPEATED])]
        Df_Subjects_dirs.append(subjects_path[0])
    
    Df_Subjects_dirs=pd.DataFrame(Df_Subjects_dirs)
    df_diff_new = pd.concat([Df_Subjects_dirs[1],next_df_par]).drop_duplicates(keep=False)
    for i in range(len(df_diff_new.index)):
        Df_Subjects_dirs=Df_Subjects_dirs.drop(df_diff_new.index[i],axis=0)
    Df_Subjects_dirs=Df_Subjects_dirs.rename(columns={0: "Path", 1: "Subject"})
    Df_Subjects_dirs = Df_Subjects_dirs.set_index('Subject')
    Def=Df_Subjects_dirs.join(result, on='Subject')
    Def.to_excel(f'{datasetDir}/CSV_DATA/Age_Path_Label_Table_{label}.xlsx')
    return Def
```

```python
def age_path_label_table(par_labels,age,label,days):
    df_par = pd.read_csv(par_labels, sep=";", header=None,skipinitialspace=True)
    df_par[0]= df_par[0].str.replace("Carlantonio", "Carlatonio", case = False) 
    df_par[0] = df_par[0].str.strip()
    df_age = pd.read_csv(age, sep=";", header=None,skipinitialspace=True)
    df_age[0] = df_age[0].str.strip()
    df_days=pd.read_csv(days, sep=";", header=None,skipinitialspace=True)
    df_days[0]= df_days[0].str.replace("Carlantonio", "Carlatonio", case = False)
    df_days[0] = df_days[0].str.strip()
    next_df_par=df_par[0]
    df_diff = pd.concat([df_par[0],df_age[0],df_days[0]]).drop_duplicates(keep=False)
    for i in range(len(df_diff.index)):
        df_age=df_age.drop([df_diff.index[i]])
    df_par.columns = ["Subject", f"{label}"]
    df_par = df_par.set_index('Subject')
    df_age.columns = ["Subject", "Age"]
    df_age = df_age.set_index('Subject')
    result = df_age.join(df_par, on='Subject')
    
    Df_Subjects_dirs = []
    for subject_path in glob.glob(f"{datasetDir}/*"):
        if subject_path.endswith(".csv"):
            continue
        if subject_path.endswith(".xlsx"):
            continue
        if subject_path.endswith("CSV_DATA"):
            continue
        subjects_path = [get_subject_metadata(subject_path , subjects_with_name=[SUBJECT_NAME for SUBJECT_NAME in SUBJECT_NAME_REPEATED])]
        Df_Subjects_dirs.append(subjects_path[0])
    
    Df_Subjects_dirs=pd.DataFrame(Df_Subjects_dirs)
    df_diff_new = pd.concat([Df_Subjects_dirs[1],next_df_par]).drop_duplicates(keep=False)
    for i in range(len(df_diff_new.index)):
        Df_Subjects_dirs=Df_Subjects_dirs.drop(df_diff_new.index[i],axis=0)
    Df_Subjects_dirs=Df_Subjects_dirs.rename(columns={0: "Path", 1: "Subject"})
    Df_Subjects_dirs = Df_Subjects_dirs.set_index('Subject')
    Def=Df_Subjects_dirs.join(result, on='Subject')
    Def.to_excel(f'{datasetDir}/CSV_DATA/Age_Path_Label_Table_{label}.xlsx')
    return Def
```
