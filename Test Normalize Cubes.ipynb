```python
import nibabel as nb
import glob as glob

for file in glob.glob("datasets/3d-dataset/*/*.nii.gz"):
    try:
        img = nb.load(file)
        data = img.get_fdata()
        
        _min = np.min(data)
        _max = np.max(data)
        if (_min == 0):
            _min_zero = np.min(data[data != 0])
            
        if _max - _min > 1:
            continue
        
        print(file)
        print("Min", _min)
        print("Max", _max)
        print("Delta", (_max - _min))
        if (_max - _min < 1):
            print("************")
        if (_min == 0):
            print("Min (exclude 0)", _min_zero)
        print("--------------------")
    except:
        print("Error in reading:", file)
        pass
```
