```python
import pandas as pd

df = pd.read_csv("dataset.csv")
```

```python
import Subject as lib
```

```python
# Local directory where the Data are mounted
rootDirectory = "/data/RMN/LUCA_PASQUINI"

# Local subdirectory where dataset is mounted
dataDir = "DATI_SEGMENTATI_SCALATI_media"

# Dataset dir
datasetDir = f"{rootDirectory}/{dataDir}"
```

```python
for subject in df.iloc[:,1]:
    
    s = lib.Subject(datasetDir, subject)
    
    try:
        t1 = s.get_sequence("T1")
        roi = s.get_roi("T2ROI")
        
        sequence_shape = (t1.shape[0], t1.shape[1])
        roi_shape = (roi.shape[0], roi.shape[1])
        
        if (sequence_shape != roi_shape):
            print(subject)
            print("\tSequence:", t1.shape)
            print("\tMask:", roi.shape)
        else:
            print(subject, "OK", sequence_shape)
            
    except:
        pass
```
