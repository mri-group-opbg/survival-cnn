# Prepare ROI sizes

```python
%%time

import glob
import os
import pickle

side=64

subjects_dir="/data/RMN2/datasets/BRATS2019/HGG_Training_data_Brats_2019"

sequence_repo = SequenceRepoBrats19(subjects_dir)

output_dir_1 = "/data/RMN/dataset-gliomi-cnn/roi-sizes-brats19"

!mkdir {output_dir_1}

for path in glob.glob(f"{subjects_dir}/HGG_Training_data_Brats_2019/*"):
    
    #print(path)
    
    subject = os.path.basename(path)
        
    #print("Working on:", subject)
    
    roi = sequence_repo.get_roi(subject)
    
    print("Working on:",subject)
    
    ((rmin, rmax), (cmin, cmax), (zmin, zmax)) = get_bounding_box(roi)
    
    z_height = zmax - zmin
    
    print(z_height)

    resampled_roi = mask_crop_resize(roi, roi, side, side, z_height)

    print("Resampled:", resampled_roi.shape)

    roi_sizes = get_roi_size_brats19(resampled_roi, 2)
    
    print(roi_sizes)

    with open(f"{output_dir_1}/roi-sizes-{subject}-{side}.pickle", "wb") as file:
            pickle.dump(roi_sizes, file)
    
    
    




```
