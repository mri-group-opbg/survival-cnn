# README

This notebook prepares a set of nii.gz cubes for signals T1, T2, FLAIR with a given fixed side of 32.
Furthermore for each subject a set of slices with the union of maximum roi where each slice has 
T1, T2 and T3 as image channels.

After reducing each signal in a cube it will be calculated the sum of sizes of cubic mask for each signal and it will be taken the set of slices with the greatest cumulative mask size

# Produce a file with ROI sizes

```python
%%time

import glob
import os
import pickle

subjects_dir = "/data/RMN/dataset-gliomi"

sequence_repo = SequenceRepoGliomi(subjects_dir)

side = 224

widths = []
heights = []

for path in glob.glob(f"{subjects_dir}/scaled/*"):
    
    subject = os.path.basename(path)
        
    print("Working on:", subject)

    try:
        roi = sequence_repo.get_roi(subject)
        
        ((rmin, rmax), (cmin, cmax), (zmin, zmax)) = get_bounding_box(roi)
        
        widths.append(rmax - rmin)
        heights.append(cmax - cmin)
        
    except:
        print("Error in roi of", subject)
        pass
```
