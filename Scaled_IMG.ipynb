```python
def normalize_name(subject_path):
    #Components= []
    subject = os.path.basename(subject_path)
    subject = re.sub(r'^(BraTS19_2013|BraTS19_CBICA|BraTS19_TCIA01|BraTS19_TCIA02|BraTS19_TCIA03|BraTS19_TCIA04|BraTS19_TCIA05|BraTS19_TCIA06|BraTS19_TCIA08|BraTS19_TMC)_', '', subject)
    components =  subject.split("_")

    return components[0]
```

```python
subjects_dir="/data/RMN2/datasets/BRATS2019/HGG_Training_data_Brats_2019/HGG_Training_data_Brats_2019"

sequence_repo = SequenceRepoBrats19(subjects_dir)

sequence_name=["t1","t1ce","t2","flair"]

output_dir = "/data/RMN2/datasets/BRATS2019/HGG_Training_data_Brats_2019/scaled"

#!mkdir {output_dir}
```

```python
for path in glob.glob(f"{subjects_dir}/*"):
    for sequence in sequence_name:
        
        subject = os.path.basename(path)
        
        roi = sequence_repo.get_roi(subject)
        
        roi_data=roi.get_fdata()
        
        img = sequence_repo.get_sequence(subject,sequence)
        
        affine=img.affine
        
        img_data=img.get_fdata()
        
        #print(area.shape, np.min(area),np.max(area))
        
        new_img=img_data/np.mean(img_data[roi_data!=0])
        
        #print(new_img.shape, np.min(new_img),np.max(new_img))
        
        new_img_nii = nb.Nifti1Image(new_img, affine)
        
        !mkdir {output_dir}/{subject}
        
        nb.save(new_img_nii, f'{output_dir}/{subject}/{subject}_{sequence}_scaled.nii')
        
        
    print("Done:",sequence)
        
        
    
    
    
    
```
