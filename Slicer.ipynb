```python
def get_mask(roi, axis, percentile):
    roi_data = roi.get_fdata()
    other_axis = list(range(len(roi_data.shape)))
    other_axis.remove(axis)
    roi_sizes = np.sum(roi_data, axis=tuple(other_axis))
    roi_indexes = np.where(roi_sizes > 0)[0]
    non_empty_sizes = roi_sizes[np.where(roi_sizes > 0)]
    percentile_val = np.percentile(non_empty_sizes, percentile)
    return np.where(roi_sizes > percentile_val)[0]
```

```python
# Plot a grid of above 70% masked data (data + mask)

import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import numpy as np

t1_orig = s.get_sequence("T1")
roi_orig = s.get_roi("T2ROI")

s = Subject(datasetDir, "ASSANTO_MARIA")

indexes = s.get_roi_index_percentile("T2ROI", 2, 70)

from nilearn.image import resample_to_img
t1_reshaped = resample_to_img(s.get_sequence("T1"), t1_orig, interpolation='nearest')
roi_reshaped = resample_to_img(s.get_roi("T2ROI"), t1_orig, interpolation='nearest')

img = t1_reshaped.dataobj
roi = roi_reshaped.dataobj

img_roi = np.asarray(img) * np.asarray(roi)

fig = plt.figure(figsize=(40., 40.))
grid = ImageGrid(fig, 111,  # similar to subplot(111)
                 nrows_ncols=(5, 4),  # creates 2x2 grid of axes
                 axes_pad=0.1,  # pad between axes in inch.
                 )

indexes = get_mask(roi_reshaped, 2, 70)

print(indexes)

for ax, im, im_roi in zip(grid, [img[:,:,z_slice] for z_slice in indexes], [img_roi[:,:,z_slice] for z_slice in indexes]):
    # Iterating over the grid returns the Axes.
    ax.imshow(im)
    ax.imshow(im_roi, alpha=0.5)

plt.show()
```
